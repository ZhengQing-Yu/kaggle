## Convert noisy text destription to a sentence embedding

Using pretrained distilbert sentence transformer

```python
import csv
import pandas as pd
import numpy as np
from scipy.spatial import distance


# pre-trained model for sentence similarity
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('distilbert-base-nli-mean-tokens')
```

```python
# shoes
sentence_a = "Nike Men Lunarswift +3 Grey Sports Shoes"
sentence_b = "Q&Q Women Vomero-5 White MM Blue Shoe"
sentence_c = "Numero Uno Men Backpack Formal Shoes"
# bottomwear
sentence_d = "John Players Flora Blue Jeans"
sentence_e = "Scullers For Mud Women White Trousers"

sentence_f, sentence_g = "Proline Nude-Coloured ColorSilk of Hilfiger T-shirts","Nike Men Cruiser Retro Black T-shirt"
sentence_h = "Mark Taylor Men Grey Striped Shirt"

sentences = [
    sentence_a,
    sentence_b,
    sentence_c,
    sentence_d,
    sentence_e,
    sentence_f,
    sentence_g,
    sentence_h
]

sentence_embeddings = model.encode(sentences)


print("shoes-shoes comparison")
print(10*(0.75 - distance.cosine(sentence_embeddings[0], sentence_embeddings[1])))
print(10*(0.75 - distance.cosine(sentence_embeddings[0], sentence_embeddings[2])))
print(10*(0.75 - distance.cosine(sentence_embeddings[1], sentence_embeddings[2])))
print("bottomwear-bottomwear comparison")
print(10*(0.75 - distance.cosine(sentence_embeddings[3], sentence_embeddings[4])))
print("shoes-bottomwear comparison")
print(10*(0.75 - distance.cosine(sentence_embeddings[0], sentence_embeddings[3])))
print(10*(0.75 - distance.cosine(sentence_embeddings[0], sentence_embeddings[4])))
print(10*(0.75 - distance.cosine(sentence_embeddings[1], sentence_embeddings[3])))
print(10*(0.75 - distance.cosine(sentence_embeddings[1], sentence_embeddings[4])))
print(10*(0.75 - distance.cosine(sentence_embeddings[3], sentence_embeddings[2])))
print("topwear-topwear comparison")
print(10*(0.75 - distance.cosine(sentence_embeddings[5], sentence_embeddings[6])))
print(10*(0.75 - distance.cosine(sentence_embeddings[5], sentence_embeddings[7])))
print(10*(0.75 - distance.cosine(sentence_embeddings[6], sentence_embeddings[7])))


# print("shoes-shoes comparison")
# print(model.get_score(sentence_a,sentence_b,metric="cosine"))
# print(model.get_score(sentence_a,sentence_c,metric="cosine"))
# print(model.get_score(sentence_b,sentence_c,metric="cosine"))

# print("bottomwear-bottomwear comparison")
# print(model.get_score(sentence_d,sentence_e,metric="cosine"))

# print("shoes-bottomwear comparison")
# print(model.get_score(sentence_a,sentence_e,metric="cosine"))
# print(model.get_score(sentence_d,sentence_b,metric="cosine"))

# print("fragrance-dress comparison")
# print(model.get_score("Park level Men Padded Deo","Gini and Jony Mania Mousse Paradise directoire Dress",metric="cosine"))

# print("fragrance-shirt comparison")
# print(model.get_score("Park level Men Padded Deo","U.S. Polo Assn. Denim Co. Men mens White Shirt",metric="cosine"))

# print("bottomwear-shirt comparison")
# print(model.get_score("Trunks Riley 66 Mumbai Diving","U.S. Polo Assn. Denim Co. Men mens White Shirt",metric="cosine"))

# print("fragrance-eyewear comparison")
# print(model.get_score("Park level Men Padded Deo","I Dee Men Black Frame Festa",metric="cosine"))

# print("topwear-topwear comparison")
# print(model.get_score("Mark Taylor Men Grey Striped Shirt","United Colors ADP1810 Benetton Women Solid Green Top",metric="cosine"))
# print(model.get_score("Mark Taylor Men Grey Striped Shirt","Nike Men Cruiser Retro Black T-shirt",metric="cosine"))
# print(model.get_score("Proline Nude-Coloured ColorSilk of Hilfiger T-shirts","Nike Men Cruiser Retro Black T-shirt",metric="cosine"))

# print("innerwear-innerwearLovable Women WTR Ciaudia Bra comparison")
# print(model.get_score("Enamor Women Weave Pink Bra","Lovable Women WTR Ciaudia Bra",metric="cosine"))
# print(model.get_score("Lovable Women WTR Ciaudia Bra","Bwitch Pink It Bra",metric="cosine"))
```

### Read data

```python
df = pd.read_csv("train.csv", delimiter=',')
df.head()
```

### Convert text into sentence embedding

```python
df['noisyTextDescription'] = df['noisyTextDescription'].map(lambda x: model.encode(x))
df.head()
```

### Write to file

```python
df.to_csv("train_preproc.csv", index=False)
```

### Do the same for test data

```python
df = pd.read_csv("test.csv", delimiter=',')
df['noisyTextDescription'] = df['noisyTextDescription'].map(lambda x: model.encode(x))
df.to_csv("test_preproc.csv", index=False)
```
